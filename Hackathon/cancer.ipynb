```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
df=pd.read_csv("dataset/cancer/Cancer_Data.csv")
df.drop(["id","Unnamed: 32"],axis=1,inplace=True)
df.tail()
```

```python
df["diagnosis"].unique()
```

```python
df["diagnosis"]=df["diagnosis"].map({"M":0,"B":1})
```

```python
col=df.columns
col
```

```python
i="fractal_dimension_worst"
df[i].unique(),df[i].min(),df[i].max()
```

```python
df["radius_mean"]=df["radius_mean"]/30
df["texture_mean"]=df["texture_mean"]/40
df["perimeter_mean"]=df["perimeter_mean"].apply(lambda x:(x-40)/(190-40))
df["area_mean"]=df["area_mean"].apply(lambda x:(x-140)/(2500-140))
df["texture_se"]=df["texture_se"]/5
df["perimeter_se"]=df["perimeter_se"]/25
df["area_se"]=df["area_se"]/550
df["radius_worst"]=df["radius_worst"]/40
df["texture_worst"]=df["texture_worst"]/50
df["perimeter_worst"]=df["perimeter_worst"].apply(lambda x:(x-50)/(250-50))
df["area_worst"]=df["area_worst"].apply(lambda x:(x-180)/(4300-180))
```

```python
df.info()
```

```python
df.head()
```

```python
x=df.drop(["diagnosis"],axis=1)
y=df["diagnosis"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python
import tensorflow as tf
```

```python
model=tf.keras.Sequential([
    tf.keras.layers.Dense(64,activation="relu"),
    tf.keras.layers.Dense(64,activation="relu"),
    tf.keras.layers.Dense(48,activation="relu"),
    tf.keras.layers.Dense(32,activation="relu"),
    tf.keras.layers.Dense(1,activation="sigmoid")
])
model.compile(optimizer=tf.optimizers.SGD(learning_rate=0.01),loss=tf.losses.binary_crossentropy,metrics=["accuracy"])
```

```python
model.fit(x_train,y_train,epochs=20,shuffle=True)
```

```python
model.evaluate(x_test,y_test)
```

```python
ypred=model.predict(x_test)
ypred=ypred.round().ravel()
confusion_matrix(y_test,ypred)
```

```python
model.save("cancerModel.h5") #0.979
```
